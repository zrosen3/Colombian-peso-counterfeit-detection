```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import common_functions as cf
from sklearn.neighbors import KNeighborsClassifier
```

# K-NN

```python
train_ds, test_ds, validation_ds, class_names = cf.LoadData()
cf.PeakData(train_ds, class_names)
```

```python
knn = KNeighborsClassifier(n_neighbors=len(class_names))
for image_batch, label_batch in train_ds:
    knn.partial_fit(image_batch, label_batch, classes=class_names)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
clf = SGDClassifier(loss='log')
for image_batch, label_batch in train_ds:
    clf.partial_fit(image_batch, label_batch, classes=class_names)
```
